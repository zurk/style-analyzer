```python
from lookout.style.format.benchmarks.compare_quality_reports import _quality_report_table_to_df

reports = {
        "v010_react_native_t": "/home/k/sourced/workdir/style-analyzer/lookout/style/format/benchmarks/reports/0.1.0/quality/react-native.train_report.md",
        "new_react_native_t": "/home/k/sourced/workdir/style-analyzer/lookout/style/format/benchmarks/reports/quality-eb2ec03d702cfbacfe2bfd1601cede2320ea4884/react-native.train_report.md",
        "v010_react_native": "/home/k/sourced/workdir/style-analyzer/lookout/style/format/benchmarks/reports/0.1.0/quality/react-native.test_report.md",
        "new_react_native": "/home/k/sourced/workdir/style-analyzer/lookout/style/format/benchmarks/reports/quality-eb2ec03d702cfbacfe2bfd1601cede2320ea4884/react-native.test_report.md",
        "v010_storybook": "/home/k/sourced/workdir/style-analyzer/lookout/style/format/benchmarks/reports/0.1.0/quality/storybook.test_report.md",
        "new_storybook": "/home/k/sourced/workdir/style-analyzer/lookout/style/format/benchmarks/reports/quality-eb2ec03d702cfbacfe2bfd1601cede2320ea4884/storybook.test_report.md",
    }

matrixes = {}
classifications = {}
    
for key in reports:
    with open(reports[key]) as f:
        content = f.read()
    content = content.split("###")
    matrix = content[2].splitlines()[2:]
    matrix = _quality_report_table_to_df(matrix)
    matrix.index = matrix.columns
    n = matrix.sum(axis=0).sum(axis=0)/100
    #matrix[]
    matrix = matrix / n
    matrixes[key] = matrix
    
    classification = _quality_report_table_to_df(content[1].splitlines()[4:])
    classification.index = classification["Class"]
    classification = classification.drop("Class", axis=1)
    classifications[key] = classification
```

```python
classifications["new_react_native"] = classifications["new_react_native"].reindex(classifications["v010_react_native"].index)
classifications["new_react_native"] = classifications["new_react_native"].fillna(0)
classifications["new_react_native"] - classifications["v010_react_native"]
```
